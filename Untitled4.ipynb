```python

from IPython import get_ipython
from IPython.display import display
# %%

# %%
class Livro:
    def __init__(self, titulo, autor, genero, quantidade):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.quantidade = quantidade

livros = []

def cadastrar_livro(titulo, autor, genero, quantidade):
    livro = Livro(titulo, autor, genero, quantidade)
    livros.append(livro)

def listar_livros():
    for livro in livros:
        print(f"Titulo: {livro.titulo}, Autor: {livro.autor}, Genero: {livro.genero}, Quantidade: {livro.quantidade}")

def buscar_livro(titulo):
    for livro in livros:
        if livro.titulo == titulo:
            return livro
    return None # Moved return None outside of the for loop so that it is only executed if the book is not found.

import matplotlib.pyplot as plt

def gerar_grafico_livros():
    generos = {}
    for livro in livros:
        if livro.genero in generos:
            generos[livro.genero] += 1
        else:
            generos[livro.genero] = 1
    plt.bar(generos.keys(), generos.values())
    plt.xlabel('Genero')
    plt.ylabel('Quantidade')
    plt.title('Quantidade de livros por Genero')
    plt.show()

# Cadastrar alguns livros
cadastrar_livro("O senhor dos anéis", "J.R.R.Tolkien", "Fantasia", 2)
cadastrar_livro("Quincas Borba", "Machado de Assis", "Romance", 4)
cadastrar_livro("O Hobbit", "J.R.R.Tolkien", "Fantasia", 6)
cadastrar_livro("Dom Casmurro", "Machado de Assis", "Romance", 4)

# Listar todos os livros
listar_livros()

# Buscar um livro pelo titulo
livro_buscado = buscar_livro("O Hobbit")
if livro_buscado:
    print(f"Livro encontrado: Titulo: {livro_buscado.titulo}, Autor: {livro_buscado.autor}, Genero: {livro_buscado.genero}, Quantidade: {livro_buscado.quantidade}")
else:
    print("Livro não encontrado")

# Gerar gráfico de quantidade de livros por genero
gerar_grafico_livros()
```
